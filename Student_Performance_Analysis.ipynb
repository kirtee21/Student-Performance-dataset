# Student Performance Analysis using Python

## Step 1: Load Dataset

In this step, the student performance dataset is loaded using the Pandas library. The dataset is stored in a DataFrame for further analysis.

## Step 2: Explore and Clean Data

In this step, the dataset is explored and cleaned by:
- Checking for missing values.
- Removing duplicate records.
- Viewing the dataset shape.
- Checking the data types of all columns.

## Step 3: Data Analysis

In this step, the dataset is analyzed to answer the following questions:
- What is the average final grade (G3)?
- How many students scored above 15?
- Is study time correlated with final grades?
- Which gender performs better on average?

## Step 4: Data Visualization

The following visualizations are created:
- Histogram of final grades (G3).
- Scatter plot showing study time versus final grades.
- Bar chart comparing the average scores of male and female students.

## Conclusion

The analysis provides insights into student performance. It shows the average final grade, the number of high-performing students, the relationship between study time and grades, and the average performance based on gender. Visualizations help in understanding the data more effectively.
